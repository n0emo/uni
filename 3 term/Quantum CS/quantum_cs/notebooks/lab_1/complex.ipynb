# Комплексные числа в Python

### Подключение необходимых библиотек.

```python
import cmath
import math
import matplotlib.pyplot as plt
```

### Создание комплексного числа.

```python
z = 23 + -5j

print(f'z = {z}')
print(f'Re(z) = {z.real}')
print(f'Im(z) = {z.imag}')
```

### Вычисление фазы и модуля комплексного числа.
Для вычисления фазы (аргумента) в модуле cmath
есть функция phase, а для вычисления модуля можно
использовать метод \_\_abs\_\_

```python
phase = cmath.phase(z)
abs = z.__abs__()

f'phase = {phase}, abs = {abs}'
```

### Изображение комплексного числа.
Для изображения применяется модуль matplotlib,
который позволяет рисовать различные графики,
в том числе в полярных координатах

```python
plt.polar([0, phase], [0, abs], marker='o')
plt.title("Число z на комплексной плоскости")

pass #plt.title возвращает результат, но выводить его не нужно
```

### Тригонометрическая форма комплексного числа.
Вычисленные фазу и модуль можно использовать для
представления числа в тригонометрической форме.

При этом, если посчитать значение этого выражение,
получится алгебраическая форма.

```python
# Тригонометрическая форма
print(f'z = {abs:.4f} * (cos({phase:.4f} + i * sin({phase:.4f}))')

# Алгебраическая форма
alg_z = abs * (math.cos(phase) + 1j * math.sin(phase))
imag_sign = '-' if alg_z .imag < 0 else '+'
print(f'z = {alg_z.real:.4f} {imag_sign} i*{math.fabs(alg_z.imag):.4f}')
```
